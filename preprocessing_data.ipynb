```python
import pandas as pd
playlistDF = pd.read_csv("data/raw_playlists.csv",index_col=False)
tracksDF=pd.read_csv("data/processed_tracks.csv")
tracksDF.head()
```

```python
tracksDF = tracksDF.drop(tracksDF.columns[0], axis=1)
playlistDF = playlistDF.drop(playlistDF.columns[0], axis=1)
tracksDF.head()
```

```python
tracksDF[['artist_name','track_name','name']]
```

```python
def drop_duplicates(df):
    '''
    Drop duplicate songs
    '''
    df['artists_song'] = df.apply(lambda row: row['artist_name']+row['track_name'],axis = 1)
    return df.drop_duplicates('artists_song')

songDF = drop_duplicates(tracksDF)
print("Todas las canciones son unicas: ",len(pd.unique(songDF.artists_song))==len(songDF))
```

```python
#seleccionando las columnas de nuestro interes
def select_cols(df):
       '''
       Select useful columns
       '''
       return df[['pos','artist_name','id','track_name','danceability',
                   'energy', 'key', 'loudness', 'mode',
                   'speechiness', 'acousticness', 'instrumentalness', 
                   'liveness', 'valence', 'tempo', "artist_pop","duration_ms_y", 
                   "genres", "track_pop"]]
songDF = select_cols(songDF)
songDF.head()
```

```python
songDF.to_csv('data/preprocessed_tracks.csv', index=False)
```
